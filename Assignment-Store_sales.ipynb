# Assignment on Store_sales dataset

**Load the Dataset**

```python
import pandas as pd
store=pd.read_csv('store_sales.csv')
store.head(10)
```

**Calculate the total sales for each store across all months.**

```python
total_sales = store.groupby('store_id')[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum()
total_sales
```

**Find the average sales for each month across all stores**

```python
total_sales = store.groupby('store_id')[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].mean()
total_sales
```

**Identify the store with the highest total sales.**

```python
total_sales = store.groupby('store_id')[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].max()
total_sales
```

**Calculate the total sales for each city.**

```python
total_city_sales = store.groupby('city')[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum()
total_city_sales
```

**List stores with total sales greater than 200**

```python
total_sales['Total'] = total_sales.sum(axis=1) 
stores_above_200 = total_sales[total_sales['Total'] > 200]
stores_above_200
```

**Which month had the highest average sales across all stores?**


```python
monthly_avg_sales = store[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].mean()
highest_avg_sales_month = monthly_avg_sales.idxmax()
highest_avg_sales_month 
```

**Which city generated the most revenue?**

```python
total_city_sales['Total'] = total_city_sales.sum(axis=1)
max_revenue_city = total_city_sales['Total'].idxmax()
max_revenue_city
```
